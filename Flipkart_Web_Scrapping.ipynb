# Importing

```python
# importing the required libraries from python 
from urllib.request import urlopen as uReq # aliasing
from bs4 import BeautifulSoup as soup
```

# HTML Upload

```python
my_url = "https://www.flipkart.com/q/best-laptops-under-rs-50000"
uclient = uReq(my_url)
page_html = uclient.read()
uclient.close()
```

# HTML parsing

```python
page_soup = soup(page_html,"html.parser")
s = soup(uclient)
```

# Read class from Flipkart Web Page

```python
containers = page_soup.findAll('div',{'class':'_2kHMtA'})
#findAll() is the most popular function in soup library

print(len(containers))
```

```python
print(soup.prettify(containers[0]))
```

```python
filename = ('best_laptops.csv')
f = open(filename,'w', encoding="utf-8")   #give encoding='utf-8' to print the results correctly
headers = "Products, Price, Processor, Ram, Size, Rating\n"
f.write(headers)
```

```python
for blocks in containers:
    products = blocks.img['alt']
    
    price_mov = blocks.findAll('div',{'class':'_30jeq3 _1_WHN1'})
    price = price_mov[0].text

    processor_mov = blocks.findAll('li',{'class':'rgWa7D'})
    processor = processor_mov[0].text     # processor info is at index 0; same class
    
    ram_mov = blocks.findAll('li',{'class':'rgWa7D'})
    ram = ram_mov[1].text                 # ram info is at index 1; same class
    

    size_mov = blocks.findAll('li',{'class':'rgWa7D'})
    size = size_mov[4].text               # screen size info is at index 5; same class
    
    try:
        rating_mov = blocks.findAll('div',{'class':'_3LWZlK'})
        rating = rating_mov[0].text
    except IndexError:
        pass

            
#     rating_mov = blocks.findAll('div',{'class':'_3LWZlK'})   # we have used try & exception above to suppress the index error through exception
#     rating = rating_mov[0].text
    
    print(products + "," + price + "," + processor + "," + ram + "," + size + "," + rating +"\n")
    f.write(products + "," + price + "," + processor + "," + ram + "," + size + "," + rating +"\n")
f.close()
```
